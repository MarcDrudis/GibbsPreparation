```python
timestep = 16
c_original = gresults[0].coriginal
prep_error = gresults[0].cfaulties[timestep] - c_original*gresults[0].betas[timestep]
errors = [np.real((g.cfaulties[timestep]-g.coriginal*g.betas[timestep])) for g in gresults]
cerrors = [ np.real(e- prep_error) for e in errors]
print([np.linalg.norm(c) for c in cerrors])
print([np.linalg.norm(e) for e in errors])
```
