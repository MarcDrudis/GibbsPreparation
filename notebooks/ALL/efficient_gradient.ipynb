```python
from qiskit.circuit import QuantumCircuit
from qiskit.circuit.library import EfficientSU2
import plotly.express as px
from qiskit.quantum_info import Operator
import numpy as np
```

```python
qc=EfficientSU2(6,entanglement="pairwise",su2_gates=[],reps=3).decompose()
qc.cx(range(0,6,2),range(1,6,2))
display(qc.draw(output="mpl"))
px.imshow(np.real(Operator(qc).data)).show()
```

```python
EfficientSU2(6,entanglement="pairwise",su2_gates=[],reps=3).decompose().draw(output="mpl")
```

```python
op = Operator(EfficientSU2(6,entanglement="reverse_linear",su2_gates=[],reps=3))
px.imshow(np.real(op.data)).show()
print(np.real(op.data))
```
